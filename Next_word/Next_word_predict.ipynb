# NEXT WORD PREDICTION

```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df = pd.read_csv("tmdb_5000_movies.csv")
df
```

```python
# to display the fist 5 value of the data set
df.head(5)
```

```python
# We need only the original title attribute
df = df["original_title"]
```

```python
df
```

```python
# we need to convert the text data into numbers for testing

# we'll convert the column from  pandas object to list type by using  

movie_name = df.to_list()
```

```python
movie_name
```

```python
# we'll use tensorflow to map each word with some unique numbers

tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(movie_name)
seq = tokenizer.texts_to_sequences(movie_name)
```

```python
seq[:10]
```

```python
# to check which number is mapped to which word we can use tokenizer.word_index

tokenizer.word_index
```

```python
# x conatins the input data
X=[]
# Y will contain the output data
Y=[]
total_words_dropped =0

# using for operation to iterate over the each word in seq

for i in seq:
    if len(i) > 1:
        for index in range(1, len(i)):
        	X.append(i[:index])
        	Y.append(i[index])
        else:
        	total_words_dropped += 1
print("The total words Dropped are :", total_words_dropped)
```

```python
X[:10]
```

```python
Y[:10]
```

```python
# the length of list is varying so wee need to make the length equal by using the keras
X = tf.keras.preprocessing.sequence.pad_sequences(X)
X
```

```python
# number of rows and column used to make the list equal
X.shape
```

```python
Y = tf.keras.utils.to_categorical(Y)
Y
```

```python
# number of rows and column used to make the list equal
Y.shape
```

```python
# vocab_size will used to store th length of the vocabulary
vocab_size = len(tokenizer.word_index) + 1
vocab_size
```

```python
# now training the model 

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 14),
    tf.keras.layers.LSTM(100, return_sequences=True),
    tf.keras.layers.LSTM(100),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(vocab_size,activation='softmax')

])
```

```python
model
```

```python
model.summary()
```

```python
model.compile(optimizer ='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,Y,epochs=10)
```

```python
model.fit(X,Y,epochs=1)
```
